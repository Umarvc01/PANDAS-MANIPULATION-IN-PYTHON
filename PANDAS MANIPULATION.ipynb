```python
SECTION B
df=pd.read_csv("C:\\Users\\USER\\Downloads\\StudentsPerformance.csv")
Test_prep=df[df["test preparation course"]=="completed"] # Selects all the students who completed the test preparation course with False= none and True= completed
print(Test_prep) 
Pass_mark=df[df["math score"]>=70] # selects students with marks of 70 and above
print(Pass_mark)
#Selecting female students with at least a score of 65 in all subjects 
Female_score= df[(df["gender"]=="female") & (df["math score"]>=65) & (df["reading score"]>=65) & (df["writing score"]>=65)]
print(Female_score)
# counts how many students have free/reduced lunch 
Free_reduced= df[df["lunch"]=="free/reduced"]
print(Free_reduced.count())

```

```python
#SECTION C
df=pd.read_csv("C:\\Users\\USER\\Downloads\\StudentsPerformance.csv")
group=df.groupby("gender")
print(group["math score"].mean()) #This is the mean math score by gender: male and female
group_parent=df.groupby("parental level of education")
print(group_parent[["math score", "reading score", "writing score"]].mean()) # all subjects by parental level of education
test_prep=df.groupby("test preparation course")
print(test_prep.mean(numeric_only=True)) # mean if students with and without test preparation course
```
